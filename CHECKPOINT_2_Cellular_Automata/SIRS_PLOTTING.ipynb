```python
import matplotlib
import sys
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import time
import scipy as sp
from scipy import ndimage

p1s=np.linspace(0,1,num=21)
p3s=np.linspace(0,1,num=21)

InfectedDataPlot = pd.read_csv("InfectedData.csv")
VarianceDataPlot=pd.read_csv("VarianceData.csv")

plot1=plt.figure(1)
plt.pcolormesh(p1s,p3s,InfectedDataPlot,cmap="inferno")
plt.title("Average Fraction of Infected Population")
plt.xlabel("Probability of Infection (p1)");
plt.ylabel("Probability of being susceptible after recovery (P3)");       
plt.colorbar()

plot2=plt.figure(2)
plt.pcolormesh(p1s,p3s,VarianceDataPlot,cmap="inferno")
plt.title("Variance")
plt.xlabel("Probability of Infection (p1)");
plt.ylabel("Probability of being susceptible after recovery (P3)"); 
plt.colorbar()


plt.show()
```
