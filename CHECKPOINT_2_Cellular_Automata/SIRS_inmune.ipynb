```python
import matplotlib
import sys
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import time
import scipy as sp
from scipy import ndimage

start_time=time.time()


#Number of Montecarlo sweeps for calculation
sweeps = 10
eqsweeps = 10

##############################################################################################################################

#Number of sweeps
N1=50

#Generate random initial state
def randominitialstate1(N1,f):   
    
    initstate = np.random.randint(-1 , 2 ,size=(N1,N1))
    
    for i in range(N1):
        for j in range(N1):
            if np.random.rand() < f:
                initstate[i,j]=2
    
    return initstate



def nextState1(cell):
    
    
    for i in range(N1):
        for j in range(N1):
            
            a = np.random.randint(0, N1)
            b = np.random.randint(0, N1)
            
            if cell[a,b] == 0:
            
                if cell[(a+1)%N1,b]==-1 or cell[a,(b+1)%N1]==-1 or cell[(a-1)%N1,b]==-1 or cell[a,(b-1)%N1]==-1:
                    if np.random.rand() < float(p1):
                        
                        cell[a,b]=-1
            
            elif cell[a,b] == -1:
                if np.random.rand() < float(p2):
                    cell[a,b] = 1
            
            elif cell[a,b] == 1:
                if np.random.rand() < float(p3):
                    cell[a,b] = 0
            
            
 
    return cell
            
def NumInfected1(cell):

    return np.count_nonzero(cell==-1)

p1=p2=p3=0.5

F=np.linspace(0,0.5,num=51)


InfePlot1=np.zeros((len(F),5))

for f in range(len(F)):

    for l in range(5):
        
    
        cell = randominitialstate1(N1,F[f])
        infe=np.zeros(sweeps)

        for i in range(eqsweeps):
            cell= nextState1(cell)

        for i in range(sweeps):         
            cell = nextState1(cell)
            ninfected=NumInfected1(cell)
           
            if ninfected != 0:
                
                infe[i]=ninfected

            else:
                infe = np.zeros(sweeps)
                break
          
        InfePlot1[f,l] = np.average(infe)       

InmuneData1 = pd.DataFrame(InfePlot1)
InmuneData1.to_csv('InmuneData1.csv',index=False)

############################################################################################################################    
#Number of sweeps
N2=100

#Generate random initial state
def randominitialstate2(N2,f):   
    
    initstate = np.random.randint(-1 , 2 ,size=(N2,N2))
    
    for i in range(N2):
        for j in range(N2):
            if np.random.rand() < f:
                initstate[i,j]=2
    
    return initstate



def nextState2(cell):
    
    
    for i in range(N2):
        for j in range(N2):
            
            a = np.random.randint(0, N2)
            b = np.random.randint(0, N2)
            
            if cell[a,b] == 0:
            
                if cell[(a+1)%N2,b]==-1 or cell[a,(b+1)%N2]==-1 or cell[(a-1)%N2,b]==-1 or cell[a,(b-1)%N2]==-1:
                    if np.random.rand() < float(p1):
                        
                        cell[a,b]=-1
            
            elif cell[a,b] == -1:
                if np.random.rand() < float(p2):
                    cell[a,b] = 1
            
            elif cell[a,b] == 1:
                if np.random.rand() < float(p3):
                    cell[a,b] = 0
            
            
 
    return cell
            
def NumInfected2(cell):

    return np.count_nonzero(cell==-1)

p1=0.8
p2=0.1
p3=0.02

F=np.linspace(0,0.5,num=51)

InfePlot2=np.zeros((len(F),5))

for f in range(len(F)):
    
    
    for l in range(5):
        
    
        cell = randominitialstate2(N2,F[f])
        infe=np.zeros(sweeps)

        for i in range(eqsweeps):
            cell= nextState2(cell)

        for i in range(sweeps):         
            cell = nextState2(cell)
            ninfected=NumInfected2(cell)
           
            if ninfected != 0:
                
                infe[i]=ninfected

            else:
                infe=np.zeros(sweeps)
                break
          
        InfePlot2[f,l] = np.average(infe)
    

InmuneData2 = pd.DataFrame(InfePlot2)
InmuneData2.to_csv('InmuneData2.csv',index=False)
       
############################################################################################################################    

InPlot1=pd.read_csv('InmuneData1.csv')
InPlot2=pd.read_csv('InmuneData2.csv')


IPlot1= InPlot1.mean(axis=1)/(N1*N1)
IPlot2= InPlot2.mean(axis=1)/(N2*N2)


#stdVar1=np.zeros(len(F))
#stdVar2=np.zeros(len(F))

#for i in range(len(F)):
        
#        shuffles1=np.zeros((1000,InPlot1.loc[[1]].size))
#        shuffles2=np.zeros((100,InPlot2.loc[[1]].size))
#        
#        I1mean=np.zeros(1000)
#        I1mean2=np.zeros(1000)
#        
#        I2mean=np.zeros(1000)
#        I2mean2=np.zeros(1000)
#        
#        Var1=np.zeros(1000)
#        Var1squared=np.zeros(1000)
#        
#        Var2=np.zeros(1000)
#        Var2squared=np.zeros(1000)
#       
#        for j in range(1000):
#            
#            for k in range(InPlot1.loc[[1]].size):
#                
#                position=int(np.random.rand()*InPlot1.loc[[1]].size)
#                shuffles1[j,k] = InPlot1.iat[i,position]
#            I1mean[j]=np.average(shuffles1[j])/(N1*N1)
#            I1mean2[j]=np.average(shuffles1[j]**2)/(N1*N1)
#            Var1[j]=I1mean2[j]-I1mean[j]**2
#            Var1squared[j]=Var1[j]**2
#            
#            for k in range(InPlot2.loc[[1]].size):
#                
#                position=int(np.random.rand()*InPlot2.loc[[1]].size)
#                shuffles2[j,k] = InPlot2.iat[i,position]
#            I2mean[j]=np.average(shuffles2[j])/(N2*N2)
#            I2mean2[j]=np.average(shuffles2[j]**2)/(N2*N2)
#            Var2[j]=I2mean2[j]-I2mean[j]**2
#            Var2squared[j]=Var2[j]**2
            
#        stdVar1[i] = math.sqrt(np.average(Var1squared)-np.average(Var1)**2)
#        stdVar2[i] = math.sqrt(np.average(Var2squared)-np.average(Var2)**2)


plot1=plt.figure(1)
plt.scatter(F,IPlot1, s=50, marker='o', color='IndianRed')
#plt.errorbar(F,IPlot1,yerr=stdVar1)
plt.title("Fraction of inmune Vs average number of infected")
plt.xlabel("Inmune fraction", fontsize=20);
plt.ylabel("Average infected fraction", fontsize=20);         plt.axis('tight');
       
plot2=plt.figure(2) 
plt.scatter(F,IPlot2, s=50, marker='o', color='IndianRed')
#plt.errorbar(F,IPlot2,yerr=stdVar2)
plt.title("Fraction of inmune Vs average number of infected")
plt.xlabel("Inmune fraction", fontsize=20);
plt.ylabel("Average infected fraction", fontsize=20);         plt.axis('tight');

print("It took: "+str((time.time()-start_time)/60)+" minutes to run")
```
